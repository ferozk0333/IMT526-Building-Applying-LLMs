```python
with open('/content/university_of_washington_overview.txt','r') as file:
  content = file.read().lower()
```

```python
content
```

```python
import re
```

```python
sentences = []
```

```python
for sentence in content.split('.'):
  if len(sentence) <5 :
    pass
  else:
    temp = re.sub('[^a-z0-9]', ' ', sentence)
    temp = re.sub('\s+', ' ', temp).strip()
    sentences.append(temp)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(sentences)
```

```python
tokenizer.word_index
```

```python
sentences[0]
```

```python
# Now we need to build a dataset: Input and Output combinations
input_sequences = []
for sentence in sentences:
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1, len(tokenized_sentence)):  # Runs n-1 times
    n_gram = tokenized_sentence[:i+1]
    input_sequences.append(n_gram)
```

```python
input_sequences
```

```python
# Now we need to make the shape of input features equal
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
input_sequences_padded = pad_sequences(input_sequences, maxlen=max_len, padding = 'pre')
```

```python
input_sequences_padded
```

```python
# Now for each array in the 2D matrix, the last element is the Output.
X = input_sequences_padded[:,:-1]
y = input_sequences_padded[:,-1]
```

```python
"""Since this is a multi-class classification task, the output for each input sequence
is a single word from a large vocabulary. We need to convert these output word indices into
One-Hot Encoded (OHE) vectors so that the model can learn to classify the correct next word
out of all possible words."""

X.shape
```

```python
y.shape
```

```python
# Let's OHE the output array which is scalar right now
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes = len(tokenizer.word_counts)+1)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
```

```python
model = Sequential()

model.add(Embedding(len(tokenizer.word_counts)+1, 100, input_length = 64))
model.add(LSTM(150))
model.add(Dense(len(tokenizer.word_counts)+1, activation = 'softmax'))
```

```python
model.build(input_shape=(None, 64))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X,y, epochs=100)
```
